```python
import numpy as np
import pandas as pd
```

```python
N = 100
COL = 2
vetor_vetores = []

x = np.random.uniform(low = -10, high = 10, size = (N, COL))
```

```python
MAX = 100
MIN = 0
birdFunc = lambda x,y: (np.sin(x)*np.square(np.exp(1-np.cos(y)))) + (np.cos(y)*np.square(np.exp(1-np.sin(y))))+np.square(x-y)

list = []
for it in x:
    list.append({
        'valor': birdFunc(it[0], it[1]),
        'x': it[0],
        'y': it[1],
    })
list.sort(key=lambda x: x['valor'], reverse=True)
list

ranked = []
for index, it in enumerate(x):
    curValue = MIN + ((MAX - MIN) * (N - index - 1) / (N-1))
    ranked.append({
        'rank': curValue,
        'object': list[index]
    })
ranked
```
